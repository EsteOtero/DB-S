```python
import json
import itertools
from flatten_dict import flatten
```

```python
with open('data.json') as json_file:
    source_data = json.load(json_file)
```

```python
def underscore_reducer(k1, k2):
    if k1 is None:
        return k2
    else:
        return '{}_{}'.format(k1,k2)
```

# Data Cleaning

```python
transactions = []
conflicts = 0 

for data_entries in itertools.islice(source_data, len(source_data)):
    
    items = data_entries['payload']['data']['socialShopping']['items']
    for item in items:
        if len(flatten(item, reducer=underscore_reducer)) != 21:
            print('1')
        transactions.append(flatten(item, reducer=underscore_reducer))
        
    
```

```python
with open('data_cleaned.json', 'w') as f:
    json.dump(transactions, f)
```

# Insert Data to DB

```python
import sqlite3
```

```python
con = sqlite3.connect("digitec.db")
c = con.cursor()
```

```python
data_to_insert = [tuple(d.values()) for d in transactions]
```

```python
c.executemany('INSERT INTO data VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)', data_to_insert)
```

```python
con.commit()
```
